```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Bengaluru_House_Data.csv')
df.head()
```

```python
## Now we will check basic info
df.info()
```

```python
## Now we will check for null values
df.isnull().sum()
```

```python
## For location only one null values so we can replace it with the available location nearby
df['location'] = df['location'].fillna(df['location'].mode()[0])
```

```python
## Now we will do value_counts for every column with help of for loop
for i in df.columns:
  print(df[i].value_counts())
```

```python
## Now we will drop some column based on null values
df.drop(['society','balcony', 'availability', 'area_type'],axis=1,inplace=True)
```

```python
df.head()
```

```python
## Now we will describe it mathematically
df.describe()
```

```python
df.info()
```

```python
## Now we will check for size and remove the null values
df['size'].value_counts()
```

```python
## As 2BHK are more so weill replace null values with it
df['size'] = df['size'].fillna('2 BHK')
```

```python
## Now for bath we will check
df['bath'].value_counts()
```

```python
## We will replace null values with median bath
df['bath'] = df['bath'].fillna(df['bath'].median())
```

```python
df.info()
```

```python
## Now we have have to find the outliers in size column
df['size'].unique()
```

```python
## We will remove the outlier having bedroom greater than 20
df['bhk'] = df['size'].str.split().str[0].astype(int)
df
```

```python
## Now we have to remove the hypen from total_sqrt column and place mean of that two values
def convert_range(x):

    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1])) / 2
    try:
        return float(x)
    except:
        return None

df['total_sqft'] = df['total_sqft'].apply(convert_range)
```

```python
## Now we will convert bedroom values in BHK form
df['size'] = df['size'].apply(lambda x: x.replace('Bedroom', 'BHK'))
df['size'].unique()
```

### Price per square feet

```python
# Now we will create price per square feet column
df['price_per_sqft'] = df['price'] * 100000 / df['total_sqft']
df.head()
```

```python
df.describe()
```

```python
# Now we will see that location whose counts are less than 10 will be replaced by 'others'
location_count = df['location'].value_counts()
location_count_less_10 = location_count[location_count <= 10]
location_count_less_10
```

```python
# Now assign these as others
df['location'] = df['location'].apply(lambda x: 'others' if x in location_count_less_10 else x)
```

```python
df['location'].value_counts()
```

## Outlier Detection and removal

```python
df.describe()
```

```python
# Here we found out that it has outlier at minimum
(df['total_sqft'] / df['bhk']).describe()
```

```python
## Now we will place condition for total_sqft/bhk >300 to remove outliers
df = df[((df['total_sqft'] / df['bhk']) >= 300)]
df.describe()
```

```python
## Now we will remove outliers from price_per_sqft and location with help of function
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        gen_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]
        df_output = pd.concat([df_output, gen_df], ignore_index=True)
    return df_output
df = remove_outliers_sqft(df)
df.describe()
```

```python
## Now we will remove the outliers from bhk with function location wize
def bhk_outlier_remover(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
            for bhk, bhk_df in location_df.groupby('bhk'):
             stats = bhk_stats.get(bhk - 1)
             if stats and stats['count'] > 5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values)
        return df.drop(exclude_indices, axis='index')
```

```python
df = bhk_outlier_remover(df)
```

```python
X = df.drop(columns=['price'])
y = df['price']
```

```python
## Now we will import train test split, regression , standardscaler, encoder, metrics , pipeline, transformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
## Now we will divide the dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Linear Regression


```python
# NOw we will use column transformer further
column_trans = make_column_transformer((OneHotEncoder(sparse_output=False), ['location']), remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
## Now we will create a pipeline for this
pipe = make_pipeline(column_trans, scaler, lr)
pipe.fit(X_train, y_train)
```

```python
## Now we will predixt for X_test
y_pred_lr = pipe.predict(X_test)
```

```python
## Now applying r2_score
r2_score(y_test, y_pred_lr)
```

## Applying Lasso

```python
lasso = Lasso()
```

```python
# NOw we create pipeline for lasso
pipe = make_pipeline(column_trans, scaler, lasso)
pipe.fit(X_train, y_train)
```

```python
# Now test the data
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test, y_pred_lasso)
```

## Applying Ridge

```python
ridge = Ridge()
```

```python
## Now we create pipeline for ridge
pipe = make_pipeline(column_trans, scaler, ridge)
pipe.fit(X_train, y_train)
```

```python
## Now test for this
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test, y_pred_ridge)
```
